```python
from sklearn import datasets
# Carrega o conjunto de dados
data = datasets.load_iris(as_frame = True)
df_iris = data.frame
df_iris.columns = ["sepal_length", "sepal_width",
"petal_length", "petal_width", "species"]
# Atualiza para os nomes das especies
df_iris["species"] = df_iris["species"].astype(str)
df_iris.loc[df_iris["species"] == "0", "species"] = "setosa"
df_iris.loc[df_iris["species"] == "1", "species"] = "versicolor"
df_iris.loc[df_iris["species"] == "2", "species"] = "virginica"
df_iris.head(4)
```

```python
df_iris.describe()
```

```python
def coef_var(df,feature):
    return df[feature].std()/df[feature].mean()

coefficient_var = pd.DataFrame([
    {"Feature":feature, "Coefficient of variation":coef_var(df_iris,feature)} for feature in df_iris.select_dtypes(include='number').columns
])
coefficient_var
```

```python
sns.boxplot(y=df_iris['sepal_length'], x = df_iris['species']);
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definir o tamanho da figura
plt.figure(figsize=(12, 8))

# Criar um gráfico para cada variável quantitativa
for i, feature in enumerate(df_iris.select_dtypes(include='number').columns, 1):
    plt.subplot(2, 2, i)  # Criar subgráficos (2 linhas, 2 colunas)
    sns.boxplot(x="species", y=feature, data=df_iris)
    plt.title(f"Distribuição de {feature} por Espécie")

plt.tight_layout()  # Ajustar os espaços
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcula a matriz de correlação
correlation_matrix = df_iris.select_dtypes(include="number").corr()

# Plota a matriz de correlação
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlação entre as Variáveis Numéricas")
plt.show()
```
